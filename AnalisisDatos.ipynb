```python
import pandas as pd
import os
```

### Cargando los archivos CSV en una lista

```python
rutaDirectorio=r"C:\Users\Alejandro\Desktop\AnalisisDatosUninorte\Data UNINORTE"
listaCSVs=[]
for archivo in os.listdir(rutaDirectorio):
    if archivo.endswith(".csv"):
        pathCompletoArchivo = os.path.join(rutaDirectorio,archivo)
        listaCSVs.append(pathCompletoArchivo)

listaCSVs[0:10]
```

### Leemos todos los archivos CSV para unirlos todos en un solo DataFrame

```python
listaDFs=[]
for csv in listaCSVs:
    dfTemp = pd.read_csv(csv,skiprows=1,delimiter=",")
    listaDFs.append(dfTemp)

listaDFs[0]
```

```python
dfDataCompleta = pd.concat(listaDFs, ignore_index=True)
dfDataCompleta
```

```python
dfDataCompleta.set_index("Rank", inplace=True)
```
